```python
import pdfplumber
import pandas
import re
import warnings
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar, LTTextBox, LTTextLine
import logging
import itertools

logging.getLogger("pdfminer.pdfinterp").setLevel(logging.ERROR)
warnings.filterwarnings("ignore")

report_name = 'FCWC Club Report.pdf'
```

```python
dados_clubes = []
start_page = 13
end_page = 82
step = 2

pages = itertools.islice(extract_pages(report_name), start_page, end_page + 1)

for i, page_layout in enumerate(pages):
  data = {}
  for element in page_layout:
    if isinstance(element, (LTTextBox, LTTextLine)):
      if 60 <= element.x0 <= 160 and 500 <= element.y0 <= 520:
        data['club'] = element.get_text().strip()
      elif 280 <= element.x0 <= 430 and 350 <= element.y0 <= 385:
        data['qualification_route'] = element.get_text().strip().replace('QUALIFICATION ROUTE', '').replace('\n', '')
      elif 280 <= element.x0 <= 340 and 310 <= element.y0 <= 340:
        data['kit_supplier'] = element.get_text().strip().replace('KIT SUPPLIER', '').replace('\n', '')
      elif 280 <= element.x0 <= 410 and 235 <= element.y0 <= 270:
        data['website'] = element.get_text().replace('\n', '').replace('WEBSITE', '').strip()
      else:
        print(f'Texto: {element.get_text().strip()}')
        print(f'Coordenadas: ({element.x0}, {element.y0}, {element.x1}, {element.y1})')
        print('---')
  # break
  if len(data) > 0:
    dados_clubes.append(data)
```
